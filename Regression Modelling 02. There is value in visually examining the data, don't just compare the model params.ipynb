## Linear Model

```r
lmX1 = lm(Y~X1, compare_df)
summary(lmX1)
```

```r
plot(Y~X1, data=compare_df)
abline(lmX1)
```

This time I am going to repeat part (a), but this time use X2 as the predictor. Do the values of s, R2, and 1*6,1 suggest a worse or better fit than the model from part (a)? Does the visual appearance of the scatterplot for the X2 model confirm or contradict the numerical findings?

```r
lmX2 = lm(Y~X2, compare_df)
summary(lmX2)

plot(Y~X2, data=compare_df)
abline(lmX2)
```

From a visual interpretations of this graph, I would suggests there is bias in theses estimates. The assumption of zero homoscedasticity (constant variance) is clearly not present in the residuals. 

(c) Repeat part (a), but this time use X3 as the predictor. Do the values of s, R2, and tb1 I suggest a worse or better fit than the model from part (a)? Does the visual appearance of the scatterplot for the X3 model confirm or contradict the numerical findings? 

```r
lmX3 = lm(Y~X3, compare_df)
summary(lmX3)

plot(Y~X3, data=compare_df)
abline(lmX3)
```

From a visual interpretations of this graph, I would suggests there is bias in theses estimates. 

Firstly the errors in the data suggest that the mean of residuals across the regression line is not equal to zero. At several points on the graph there is a clear skew, with the residuals in the left and right corners of the graph mostly beneath the regression line. This suggests the zero mean assumption is not upheld.

The second assumption that may be broken here is the constant variance, again we see a large difference in the width of residual error between sections of the plot.

(d) Repeat part (a), but this time use X4 as the predictor. Do the values of s, R2, and tb1 I suggest a worse or better fit than the model from part (a)? Does the visual appearance of the scatterplot for the X4 model confirm or contradict the numerical findings?

```r
lmX4 = lm(Y~X4, compare_df)
summary(lmX4)

plot(Y~X4, data=compare_df)
abline(lmX4)
```
