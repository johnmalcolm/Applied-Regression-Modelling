# Regression Modelling 6
## Transformations into Quadratic and Square Root Models

```r
install.packages("tidyverse")
install.packages("gridExtra")
install.packages("nortest")

require(tidyverse)
require(gridExtra)
require("nortest")
options(repr.plot.width=20, repr.plot.height=9)

nyjuice_df <- read.table(url("http://jupyter-regressions.s3.eu-west-1.amazonaws.com/nyjuice.csv"), 
                           sep=",", header = T, stringsAsFactors = F)
```

```r
head(nyjuice_df)
```

```r
plot(nyjuice_df)
```

## Simple Model

```r
simple.model <-lm(Cases ~ Day, data=nyjuice_df)
summary(simple.model)

ggplot(nyjuice_df, aes(y=Cases, x=Day)) + 
  geom_point(alpha = .5) + 
  stat_smooth(method = "lm", formula = y ~ x)
```

## Quadratic Model

```r
quad.model <-lm(Cases ~ Day + I(Day^2), data=nyjuice_df)
summary(quad.model)

ggplot(nyjuice_df, aes(y=Cases, x=Day)) + 
  geom_point(alpha = .5) + 
  stat_smooth(method = "lm", formula = y ~ x+I(x^2))
```

## Square Root Model

```r
sqrt.model <-lm(Cases~ Day + I(Day^0.5), data=nyjuice_df)
summary(sqrt.model)

ggplot(nyjuice_df, aes(y=Cases, x=Day)) + 
  geom_point(alpha = .5) + 
  stat_smooth(method = "lm", formula = y ~ x+I(x^0.5))
```
